```python
def con_st(x):
    return '{:02}'.format(x)
```

```python
def con_dist(x):
    return '{:02}'.format(x)
```

```python
def con_teh(x):
    return '{:04}'.format(x)
```

```python
def con_vil(x):
    x = str(x)
    req_z = 8 - len(x)
    ans = ('0'*req_z) + x
    return ans
```

```python
asset_data_2001['ST_CODE'] = asset_data_2001['ST_CODE'].map(con_st)
```

```python
asset_data_2001['DIST_CODE'] = asset_data_2001['DIST_CODE'].map(con_dist)
```

```python
asset_data_2001['THSIL_CODE'] = asset_data_2001['THSIL_CODE'].map(con_teh)
```

```python
asset_data_2001['V_CT_CODE'] = asset_data_2001['V_CT_CODE'].map(con_vil)
```
